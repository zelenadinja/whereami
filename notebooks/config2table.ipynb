```python
import pandas as pd 
import df2img
from src.utils import read_artifacts_s3
```

```python
config = read_artifacts_s3('config_files/version_0.yaml')
```

```python
df = pd.DataFrame.from_dict(config, orient='index', columns=['params'])
```

```python
import df2img

fig = df2img.plot_dataframe(
    df,
    title=dict(
        font_color="darkred",
        font_family="Times New Roman",
        font_size=30,
        text="Config for Version 0",
    ),
    tbl_header=dict(
        align="right",
        fill_color="black",
        font_color="white",
        font_size=20,
        line_color="darkslategray",
    ),
    tbl_cells=dict(
        align="left",
        font_size=10,
        line_color="darkslategray",
    ),
    row_fill_color=("#ffffff", "#d7d8d6"),
    fig_size=(500, 500),
)

df2img.save_dataframe(fig=fig, filename="plot.png")
```
